## Histogramme d'une image
Utiliser [numpy.histogram](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.histogram.html) pour calculer l'historamme des niveaux de gris et [plt.bar](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html) pour l'afficher.

Combien de pixels de l'image sont-ils noirs ? 
(Question bonus: Vérifier ce résultat de deux façons différentes, à l'aide d'une [sélection](https://thispointer.com/python-numpy-select-elements-or-indices-by-conditions-from-numpy-array/).

```python
# np.unique(img_gray)
```

```python
# Réponse
hist, bins = np.histogram(img_gray, bins=10)
width = (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
print(width, center)
plt.bar(center, hist, align='center', width=width)
plt.show()
```
