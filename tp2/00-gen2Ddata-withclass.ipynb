# Génération d'un dataset bidimensionnel avec scikit-learn
et création d'un fichier CSV

Références: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html
et https://machinelearningmastery.com/generate-test-datasets-python-scikit-learn/

Identique à gen2Ddata, mais enregistre $y$ dans la troisième colonne.


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs

# from sklearn.datasets.samples_generator import make_blobs

X, y = make_blobs(n_samples=200, centers=4, n_features=2, random_state=123)
print(X.shape)
print(y.shape)
```

```python
# Plot
fig, ax = plt.subplots()
df = pd.DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue', 2:'green', 3:'grey'}
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])

# plt.savefig('kms-dataset2d.pdf')
plt.show()
```

```python
# Export en fichier CSV
# Ajoute y en troisième colonne
XY = np.concatenate( (X, y.reshape((-1,1))), axis=1 )
np.savetxt("./kms-dataset2d-XY.csv", XY, delimiter=";")
print("Matrice de taille", XY.shape, "enregistrée")
```
